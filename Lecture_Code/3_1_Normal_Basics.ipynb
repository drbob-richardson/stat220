Plot a normal density function. Play around with the mean and sd to get different curves

```python
# importing libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
  
# Plot between -20 and 20 with .01 steps.
x_axis = np.arange(start = -20, stop = 20, step = 0.01)

# Calculating mean and standard deviation
mean = -12
sd = .5
  
plt.plot(x_axis, norm.pdf(x_axis, mean, sd))
plt.show()
```

Draw realizations from a normal distribution and plot as a histogram. Change location and scale to get different shapes.

```python
# take 1000 draws where loc is the mean and scale is the sd
draws = np.random.normal(size = 1000, loc = 0, scale = 1)

# Plot histogram of the resulting draws
# can use plt.hist(draws), other options just my personal taste
plt.hist(draws, bins = 30,alpha=.7,rwidth = 0.9) 
plt.show()
```

Plot normal draws as a density and overlay theoretical normal plot

```python
# mean and sd values
mean = 0
sd = 100

# Draw 1000 random normals
draws = np.random.normal(size = 10000, loc = mean, scale = sd)

# Plot histogram with density =True to scale to be a density and not a frequency
plt.hist(draws, bins = 30,alpha=.7,rwidth = 0.9,density=True)

# Overlay theoretical curve
x_axis = np.arange(start = -400, stop = 400, step = 0.01)
plt.plot(x_axis, norm.pdf(x_axis, mean, sd))
plt.show()
```
