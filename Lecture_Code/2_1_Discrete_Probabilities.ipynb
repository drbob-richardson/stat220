Loading required packages

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import binom, geom
import scipy.stats as stats
```

This code generates 1000 random draws from a binomial distribution with n = 10 and p = 0.3 and plots all the possible results in a histogram

```python
# Parameters for the binomial distribution
n = 10  # Number of trials
p = 0.3  # Probability of success in each trial

# Generate random samples from the binomial distribution
binom_samples = binom.rvs(n=n, p=p, size=1000)

# Plot the probability mass functions (PMFs) for binomial and geometric distributions
plt.figure(figsize=(12, 5))

plt.hist(binom_samples, bins=np.arange(0, n + 2) - 0.5, density=True, alpha=0.6, color='blue', edgecolor='black')
plt.title('Binomial Distribution (n=10, p=0.3)')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.grid(True)
```

This code generates 1000 random draws from a geometric distribution with p = 0.3 and plots them as a histogram

```python


# Parameters for the geometric distribution
p_geom = 0.3  # Probability of success in each trial for geometric distribution

# Generate random samples from the geometric distribution
geom_samples = geom.rvs(p=p_geom, size=100000)

# Plot the probability mass functions (PMFs) for binomial and geometric distributions
plt.figure(figsize=(12, 5))

plt.hist(geom_samples, bins=np.arange(0, max(geom_samples) + 2) - 0.5, density=True, alpha=0.6, color='green', edgecolor='black')
plt.title('Geometric Distribution (p=0.3)')
plt.xlabel('Number of Trials Until First Success')
plt.ylabel('Probability')
plt.grid(True)

```

We can use python functions to find exact Binomial probabilities or even to find ranges. The stats.binom.cdf function calculates the sum of probabilities from 0 to k, so we can find the probability of k or less successes.

```python
# Calculate probabilities for specific events in the binomial distribution
binom_prob_success_3 = stats.binom.pmf(3, n, p)  # Probability of getting 3 successes in 10 trials
binom_prob_leq_5 = stats.binom.cdf(5, n, p)  # Probability of 5 or less successes in 10 trials

# Print the results
print(f"Binomial Distribution:")
print(f"Probability of 3 successes in 10 trials: {binom_prob_success_3:.4f}")
print(f"Probability of 5 or less successes in 10 trials: {binom_prob_leq_5:.4f}")
```

We can find exact probabilties for the geometric distribution as well as the sum of probabilities between 0 and k, which is success at least by the k-th trial.

```python

# Calculate probabilities for specific events in the geometric distribution
geom_prob_success_on_4th_trial = stats.geom.pmf(3, p_geom)  # Probability of success on the 4th trial
geom_prob_success_by_5th_trial = stats.geom.cdf(5, p_geom)  # Probability of success by the 5th trial


print(f"\nGeometric Distribution:")
print(f"Probability of success on the 4th trial: {geom_prob_success_on_4th_trial:.4f}")
print(f"Probability of success by the 5th trial: {geom_prob_success_by_5th_trial:.4f}")
```
