# Continuous Probability Workbook
A guided notebook for exploring Normal, Uniform, and Exponential distributions.

**Goals:**
- Visualize PDFs and CDFs
- Compute probabilities with SciPy
- Find quantiles and percentiles
- Practice applied examples

## Setup

```python

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, uniform, expon

def plot_xy(x, y, title, xlabel="x", ylabel="f(x)"):
    plt.figure()
    plt.plot(x, y)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()

print("Libraries ready.")
```

## Normal Distribution
Definition, visualization, probabilities, quantiles.

```python

mu, sigma = 70, 3
x = np.linspace(mu-4*sigma, mu+4*sigma, 500)
pdf = norm.pdf(x, mu, sigma)
cdf = norm.cdf(x, mu, sigma)
plot_xy(x, pdf, f"Normal PDF (mu={mu}, sigma={sigma})")
plot_xy(x, cdf, f"Normal CDF (mu={mu}, sigma={sigma})", ylabel="F(x)")
```

### Normal: Probability example

```python

a, b = 67, 73
p = norm.cdf(b, mu, sigma) - norm.cdf(a, mu, sigma)
print(f"P({a} <= X <= {b}) = {p:.4f}")
```

### Normal: Quantile example

```python

q30 = norm.ppf(0.30, mu, sigma)
print("30th percentile =", q30)
```

## Uniform Distribution
Definition, visualization, probabilities, quantiles.

```python

a, b = 0, 2
x = np.linspace(-0.5, 2.5, 400)
pdf = uniform.pdf(x, a, b-a)
cdf = uniform.cdf(x, a, b-a)
plot_xy(x, pdf, f"Uniform PDF [{a},{b}]")
plot_xy(x, cdf, f"Uniform CDF [{a},{b}]", ylabel="F(x)")
```

### Uniform: Probability & Quantile

```python

p = uniform.cdf(1.5, a, b-a) - uniform.cdf(0.5, a, b-a)
q30 = uniform.ppf(0.30, a, b-a)
print("P(0.5 <= X <= 1.5) =", p)
print("30th percentile =", q30)
```

## Exponential Distribution
Definition, visualization, probabilities, quantiles.

```python

lam = 0.5
x = np.linspace(0, 10, 400)
pdf = expon.pdf(x, scale=1/lam)
cdf = expon.cdf(x, scale=1/lam)
plot_xy(x, pdf, f"Exponential PDF (lambda={lam})")
plot_xy(x, cdf, f"Exponential CDF (lambda={lam})", ylabel="F(x)")
```

### Exponential: Probability & Quantile

```python

p = expon.cdf(3, scale=1/lam) - expon.cdf(1, scale=1/lam)
q80 = expon.ppf(0.80, scale=1/lam)
print("P(1 <= X <= 3) =", p)
print("80th percentile =", q80)
```

## Applied Examples

### Dinner Time (Uniform)
Dinner is equally likely between 5:007:00 pm (2 hours).

```python

a, b = 0, 2
p = uniform.cdf(1.5, a, b-a) - uniform.cdf(0.5, a, b-a)
q30 = uniform.ppf(0.30, a, b-a)
print("Mean dinner time (hours past 5) =", (a+b)/2)
print("P(0.5 <= X <= 1.5) =", p)
print("30th percentile =", q30)
```

### Heights (Normal)
Heights $X \sim N(70,3)$.

```python

mu, sigma = 70, 3
p = norm.cdf(73, mu, sigma) - norm.cdf(67, mu, sigma)
q30 = norm.ppf(0.30, mu, sigma)
print("P(67 <= X <= 73) =", p)
print("30th percentile =", q30)
```

### Bus Wait (Exponential)
Wait time $X \sim Exp(\lambda=1/10)$ (mean=10 min).

```python

lam = 1/10
p = expon.cdf(15, scale=1/lam) - expon.cdf(5, scale=1/lam)
q80 = expon.ppf(0.80, scale=1/lam)
print("P(5 <= X <= 15) =", p)
print("80th percentile =", q80)
```
