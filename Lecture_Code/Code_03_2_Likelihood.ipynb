Plot the likelihood function for data [1, 2, 3, 4, 5]. Change the data to see different results. We will assume that mu is unknown but sigma is fixed at 1.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import norm

# Get data ready
data = [1, 2, 3, 4, 5]

# Plot between 1 and 5 with .01 steps.
x_axis = np.arange(start=1, stop=5, step=0.01)

# Find the likelihood for different values of mu
likelihood_values = [np.prod(norm.pdf(data, mu, 1)) for mu in x_axis]

# Create a new figure with a specific size
plt.figure(figsize=(6, 3))

# Plot the likelihood values with improved aesthetics
plt.plot(x_axis, likelihood_values, color='darkblue', linewidth=2)

# Adding grid lines for better readability
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Adding title and labels with increased font size for clarity
plt.title('Likelihood Values for Different Mu', fontsize=16)
plt.xlabel('Mu Values', fontsize=14)
plt.ylabel('Likelihood', fontsize=14)

# Marking the maximum likelihood value with a red dot and text annotation
max_likelihood_mu = x_axis[np.argmax(likelihood_values)]
max_likelihood_value = max(likelihood_values)
plt.plot(max_likelihood_mu, max_likelihood_value, marker='o', markersize=5, color="red")
plt.text(max_likelihood_mu, max_likelihood_value, f'Max Likelihood\nMu = {max_likelihood_mu:.2f}', fontsize=12, ha='right')

# Adjust the y-axis limit to be one unit higher than the maximum likelihood value
plt.ylim(0, max_likelihood_value *1.2)

# Show the enhanced plot
plt.show()

```

Assume mu is equal to the mean of the data and sigma is unknown

```python

# Get data ready
data = [1, 2, 3, 4, 5]

# Plot between 0.01 and 5 with .01 steps.
x_axis = np.arange(start=0.01, stop=5, step=0.01)

# Create a new python function that calculates the likelihood
likelihood_values = [np.prod(norm.pdf(data, 3, sig)) for sig in x_axis]

# Create a new figure with a specific size
plt.figure(figsize=(6, 3))

# Plot the likelihood values with improved aesthetics
plt.plot(x_axis, likelihood_values, color='darkblue', linewidth=2)

# Adding grid lines for better readability
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Adding title and labels with increased font size for clarity
plt.title('Likelihood Values for Different Sigma', fontsize=16)
plt.xlabel('Sigma Values', fontsize=14)
plt.ylabel('Likelihood', fontsize=14)

# Marking the maximum likelihood value with a red dot and text annotation
max_likelihood_sig = x_axis[np.argmax(likelihood_values)]
max_likelihood_value = max(likelihood_values)
plt.plot(max_likelihood_sig, max_likelihood_value, marker='o', markersize=5, color="red")
plt.text(max_likelihood_sig, max_likelihood_value, f'Max Likelihood\nSigma = {max_likelihood_sig:.2f}', fontsize=12, ha='right')

# Adjust the y-axis limit to be one unit higher than the maximum likelihood value
plt.ylim(0, max_likelihood_value*1.2)

# Show the enhanced plot
plt.show()
```

Find exact values using MLE formulas we learned in class

```python

# Calculate the MLEs
mu_mle = np.sum(data)/len(data)
sigma_mle = np.sqrt(np.sum((data-mu_mle)**2)/len(data))

# Create a DataFrame to display the results nicely
results_df = pd.DataFrame({
    "Parameter": ["Mu MLE", "Sigma MLE"],
    "Value": [mu_mle, sigma_mle]
})

# Display the DataFrame
print(results_df)

```
