```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler
```

New Data set! This is data on Indian youtubers and viewers were given a chance to rate how much they trust the content in the videos they release. The trust rating is the variable "Trustwortiness" and it is ona scale of 1 to 10. SME is Source material expertise. There is a variable for if the video is one that required souce material expertise or if it did not and a variable for if the creator had the source material expertise. Experience is the time since creation of the channel to release of the video in days. Many of the other variables are labeled in such a way to understand what they contain.

```python
trust = pd.read_csv("https://richardson.byu.edu/220/trust.csv",encoding_errors='ignore')
trust
```

```python
features = trust[['Channel_Subscribers', 'TotalViewsoftheChannel', 'Experience','SME_Creator']]
df_encoded = pd.get_dummies(features, drop_first=True)

# Selecting features and target variable
target = trust['Trustworthiness']
df_encoded
```

Fit a model for trust based on some other values in the model

```python
features = trust[['Channel_Subscribers', 'TotalViewsoftheChannel', 'Experience','SME_Creator']]
features_encoded = pd.get_dummies(features, drop_first=True)

# Selecting features and target variable
target = trust['Trustworthiness']

# Splitting the dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(features_encoded, target, test_size=0.2, random_state=42)

# Linear Regression on original data
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)

# Predicting on the test set
y_pred = lin_reg.predict(X_test)

# Calculating R^2 value on the test set
r2_original = r2_score(y_test, y_pred)
print(f'R^2 value on the original data: {r2_original}')
```

The R^2 you are seeing is very small, but it might not be trusted because when you work with really large or small values some numerical issues may occur. How small are the coefficients we get?

```python
coefficients = lin_reg.coef_
intercept = lin_reg.intercept_

# Printing the coefficients and the intercept
print(f'Intercept: {intercept}')
print('Coefficients:')
for feature, coef in zip(features.columns, coefficients):
    print(f'{feature}: {coef}')
```

```python
Xscaler = StandardScaler()
X_train_scaled = pd.DataFrame(Xscaler.fit_transform(X_train),columns = X_train.columns.to_list())

X_test_scaled = pd.DataFrame(Xscaler.transform(X_test),columns = X_train.columns.to_list())

yscaler = StandardScaler()
y_train_scaled = yscaler.fit_transform(y_train.values.reshape(-1, 1))
y_test_scaled = yscaler.transform(y_test.values.reshape(-1, 1))

X_train_scaled
```

We fit the model on the scaled data

```python
# Fit the model on the scaled training data
model = LinearRegression()
model.fit(X_train_scaled, y_train_scaled)

# Predict on the scaled training data
y_train_pred_scaled = model.predict(X_train_scaled)

# Calculate the R^2 score on the scaled training data predictions
r2_train_scaled = r2_score(y_train_scaled, y_train_pred_scaled)
print(f'R^2 for the scaled training data: {r2_train_scaled}')
```

Let's back transform to the original scale.

```python
# To use the R^2 score for the original scale, inverse-transform the predictions
y_train_pred = yscaler.inverse_transform(y_train_pred_scaled.reshape(-1, 1))
r2_train_original_scale = r2_score(y_train, y_train_pred)
print(f'R^2 for the training data on the original scale: {r2_train_original_scale}')
```

## Important Note!!!
When comparing models with scaled and standardized data
- If there are no computational issues, R^2 will be the same for both models
- MSE and rMSE should not be used to compare between the models directly, because we transformed the target variable
