1. Let X be a normal random variable. Ten realizations of the random variable have a mean of 5. What can you infer about the variance of X?

2. Provide code that draws 100 random values from a normal distribution with $\mu = 100$ and $\sigma = 100$.

3. Provide code that plots a normal probability density function with a mean of -5 and a standard deviation of 1.

4. Assume that data comes from a normal distribution with mean  and variance . The observations are [1, 2, 2, 6, 7, 11, 12]. What is the MLE estimate for $\mu$?

5. Using the data from problem 4, what is the MLE estimate of $\sigma^2$?


6. Draw 1000 random variables from a normal distribution with a mean of 50 and a variance of 100. Create an image of a histogram of the sampled data with the theoretical distribution overlaid.

7. Assume that data comes from a normal distribution with mean $\mu$ and a standard deviation of 5. The observations are [1, 2, 2, 2, 3, 3, 3, 3, 10, 15, 20]. Plot the likelihood function as a function of  over the range of the data.

8. Using online resources look up the following distributions:

  (a) Poisson
  (b) gamma
  (c) log-normal
  (d) Cauchy
  (e) beta
  (f) exponential

For each distribution, write down how many parameters the distribution has and what is it's domain, meaning what values can it provide probability for. For example, the normal distribution is for values between $-\infty$ and $\infty$ but the Bernoulli is for the set $\{0,1\}$.
